```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def EV1(udt):
    udt = np.copy(udt)
    udt[:] = 1
    return udt
def EV2(udt):
    return udt - np.sqrt(udt**2 + 4)
def EV3(udt):
    return udt + np.sqrt(udt**2 + 4)

udt = np.linspace(-10, 10, 1000)
EV1_y = EV1(udt)
EV2_y = EV2(udt)
EV3_y = EV3(udt)

plt.plot(udt, EV1_y, c = 'midnightblue', label = r'$EV_1$')
plt.plot(udt, EV2_y, c = 'firebrick', label = r'$EV_2$')
plt.plot(udt, EV3_y, c = 'forestgreen', label = r'$EV_3$')
plt.xlabel(r'$t$')
plt.ylabel(r'$E / t$')
plt.grid()
plt.legend()
plt.savefig('plot.pdf')
None
```
