```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def EV1(udt):
    udt = np.copy(udt)
    return udt

def EV2(udt):
    return udt/2 + np.sqrt(1/4*udt**2 + 4)
def EV3(udt):
    return udt/2 - np.sqrt(1/4*udt**2 + 4)

def EV4(udt):
    udt = np.copy(udt[:])
    udt[:] = 0
    return udt

udt = np.linspace(-5, 5, 1000)
EV1_y = EV1(udt)
EV2_y = EV2(udt)
EV3_y = EV3(udt)
EV4_y = EV4(udt)


plt.plot(udt, EV1_y, c = 'midnightblue', label = r'$EV_1$')
plt.plot(udt, EV2_y, c = 'firebrick', label = r'$EV_2$')
plt.plot(udt, EV3_y, c = 'forestgreen', label = r'$EV_3$')
plt.plot(udt, EV4_y, c = 'orange', label = r'$EV_4$')

plt.xlabel(r'$U / t$')
plt.ylabel(r'$E / t$')
plt.grid()
plt.legend()
plt.savefig('plot.pdf')
None
```
